```python
for i in range(6):
    df = pd.read_csv(f'/home/chanapa/BCI_fork/BCI/P300/th_data/sit_cb/sit_cb{i}.csv')
    df = df.drop(["timestamps"], axis=1)

    raw = helper.df_to_raw(df)
    raw.notch_filter(np.arange(50, 125, 50), filter_length='auto', phase='zero') #250/2 based on Nyquist Theorem
    raw.filter(1, 20, method='iir')

    event_id = {'Non-Target': 1, 'Target' : 2}

    tmin = 0.2
    tmax = 0.5
    picks = 'eeg'
    epochs = helper.getEpochs(raw, event_id, tmin, tmax, picks)

    if i == 0 :
        X = epochs.get_data()
        y = epochs.events[:, -1]
    else :
        X = np.concatenate((X, epochs.get_data()), axis = 0 )
        y = np.concatenate((y, epochs.events[:, -1]), axis = 0 )
    
    print(X.shape)
    print(y.shape)
```
